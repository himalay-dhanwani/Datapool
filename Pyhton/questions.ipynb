```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = pd.read_csv("/content/project datapool_df.csv")
```

1. What are the top-performing product categories?

```python
# Group by Product Category and sum Net Amount to find top-performing categories
top_categories = df.groupby('Product Category')['Net Amount'].sum().sort_values(ascending=False)

# Plotting the top product categories
plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.index, y=top_categories.values, palette='viridis')
plt.title('Top Performing Product Categories')
plt.xticks(rotation=45)
plt.ylabel('Total Net Amount')
plt.show()
```

2. How does discount usage affect the Net Amount or Gross Amount?

```python
# Relationship between Discount Availed and Net Amount
plt.figure(figsize=(10,6))
sns.boxplot(x='Discount Availed', y='Net Amount', data=df, palette='coolwarm')
plt.title('Impact of Discount Usage on Net Amount')
plt.show()

# Relationship between Discount Availed and Gross Amount
plt.figure(figsize=(10,6))
sns.boxplot(x='Discount Availed', y='Gross Amount', data=df, palette='coolwarm')
plt.title('Impact of Discount Usage on Gross Amount')
plt.show()

```

3. Which gender or age group is the most active in terms of purchase volume?

```python
# Grouping by Gender and Age Group for total purchases
purchase_volume_gender = df.groupby('Gender')['TID'].count().sort_values(ascending=False)
purchase_volume_age = df.groupby('Age Group')['TID'].count().sort_values(ascending=False)

# Plotting Gender-based purchase volume
plt.figure(figsize=(8,5))
sns.barplot(x=purchase_volume_gender.index, y=purchase_volume_gender.values, palette='Set2')
plt.title('Purchase Volume by Gender')
plt.ylabel('Number of Purchases')
plt.show()

# Plotting Age-based purchase volume
plt.figure(figsize=(10,6))
sns.barplot(x=purchase_volume_age.index, y=purchase_volume_age.values, palette='Set2')
plt.title('Purchase Volume by Age Group')
plt.ylabel('Number of Purchases')
plt.show()

```

4. How does location influence product preferences and sales performance?

```python
# Grouping by Location and Product Category for sales
location_product_performance = df.groupby(['Location', 'Product Category'])['Net Amount'].sum().unstack()

# Heatmap to show performance by location and product category
plt.figure(figsize=(12,8))
sns.heatmap(location_product_performance, annot=True, cmap='Blues')
plt.title('Sales Performance by Location and Product Category')
plt.show()
```

5. What are the peak sales periods based on Purchase Date (monthly/seasonal trends)?

```python
# Convert Purchase Date to datetime format
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])

# Extracting month and year from Purchase Date
df['Month'] = df['Purchase Date'].dt.month
df['Year'] = df['Purchase Date'].dt.year

# Group by Month and sum Net Amount
monthly_sales = df.groupby('Month')['Net Amount'].sum()

# Plotting monthly sales trends
plt.figure(figsize=(10,6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Monthly Sales Trends')
plt.ylabel('Total Net Amount')
plt.xlabel('Month')
plt.show()
```

6. How does the Net Amount vary between different purchase methods (online vs. in-store)?

```python
# Group by Purchase Method and sum Net Amount
purchase_method_sales = df.groupby('Purchase Method')['Net Amount'].sum()

# Plotting sales performance by purchase method
plt.figure(figsize=(8,5))
sns.barplot(x=purchase_method_sales.index, y=purchase_method_sales.values, palette='coolwarm')
plt.title('Net Amount by Purchase Method')
plt.ylabel('Total Net Amount')
plt.show()
```

7. What impact do specific discount names have on sales performance?

```python
# Group by Discount Name and sum Net Amount
discount_performance = df.groupby('Discount Name')['Net Amount'].sum()

# Plotting sales performance by Discount Name
plt.figure(figsize=(10,6))
sns.barplot(x=discount_performance.index, y=discount_performance.values, palette='Set3')
plt.title('Sales Performance by Discount Name')
plt.xticks(rotation=45)
plt.ylabel('Total Net Amount')
plt.show()
```

8. Which regions are benefiting most from discounts?

```python
# Filter rows where Discount was availed
discount_df = df[df['Discount Availed'] == 'Yes']

# Group by Location and sum Discount Amount
discount_by_location = discount_df.groupby('Location')['Discount Amount (INR)'].sum()

# Plotting discount usage by location
plt.figure(figsize=(10,6))
sns.barplot(x=discount_by_location.index, y=discount_by_location.values, palette='coolwarm')
plt.title('Discounts Availed by Location')
plt.xticks(rotation=45)
plt.ylabel('Total Discount Amount (INR)')
plt.show()
```

9. How does the average order value differ between first-time and repeat customers?

```python
# Assuming 'TID' indicates whether a customer is first-time or repeat (unique vs multiple TIDs per CID)
customer_order_count = df.groupby('CID')['TID'].count()

# Classify customers as First-Time or Repeat
df['Customer Type'] = df['CID'].apply(lambda x: 'Repeat' if customer_order_count[x] > 1 else 'First-Time')

# Calculate average order value for first-time and repeat customers
avg_order_value = df.groupby('Customer Type')['Net Amount'].mean()

# Plotting average order value
plt.figure(figsize=(8,5))
sns.barplot(x=avg_order_value.index, y=avg_order_value.values, palette='Set2')
plt.title('Average Order Value: First-Time vs Repeat Customers')
plt.ylabel('Average Net Amount')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (replace with your actual dataset path)
data = pd.read_csv("/content/project datapool_df.csv")

# Filter data for customers from Delhi
delhi_customers = data[data['Location'] == 'Delhi']

# Debugging output for filtered data
print("Filtered Delhi Customers:\n", delhi_customers.head())
print("Number of Delhi Customers:", delhi_customers.shape[0])

# Group data by customer ID to calculate loyalty metrics
customer_loyalty = delhi_customers.groupby('CID').agg({
    'TID': 'count',                      # Number of transactions (frequency)
    'Net Amount': 'sum',                 # Total purchase value
    'Purchase Date': 'max'               # Most recent purchase date
}).reset_index()

# Rename columns for clarity
customer_loyalty.columns = ['Customer ID', 'Purchase Frequency', 'Total Spend', 'Last Purchase Date']

# Convert 'Last Purchase Date' to datetime to calculate recency (optional)
customer_loyalty['Last Purchase Date'] = pd.to_datetime(customer_loyalty['Last Purchase Date'], dayfirst=True)

# Define thresholds or scores for loyalty
def loyalty_score(row):
    if row['Purchase Frequency'] > 3 and row['Total Spend'] > 5000:
        return 'High'
    elif row['Purchase Frequency'] > 1 and row['Total Spend'] > 2000:
        return 'Medium'
    else:
        return 'Low'

# Apply the loyalty score function
customer_loyalty['Loyalty Score'] = customer_loyalty.apply(loyalty_score, axis=1)

# Create a pivot table for heatmap visualization
heatmap_data = customer_loyalty.pivot_table(
    index='Purchase Frequency',
    columns='Loyalty Score',
    values='Total Spend',
    aggfunc='sum',
    fill_value=0
)

# Debugging output for heatmap data
print("Customer Loyalty Data:\n", customer_loyalty)
print("Heatmap Data:\n", heatmap_data)

# Plotting the heatmap if data is not empty
if heatmap_data.empty:
    print("Heatmap data is empty. No data to plot.")
else:
    plt.figure(figsize=(10, 6))
    sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")
    plt.title('Total Spend by Purchase Frequency and Loyalty Score (Delhi Customers)')
    plt.xlabel('Loyalty Score')
    plt.ylabel('Purchase Frequency')
    plt.show()

# Optional: Display the customer loyalty metrics DataFrame
print(customer_loyalty)

# Optional: Export the results to a CSV file
customer_loyalty.to_csv('customer_loyalty_delhi.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (replace with your actual dataset path)
data = pd.read_csv("/content/project datapool_df.csv")

# Group data by customer ID to calculate loyalty metrics
customer_loyalty = data.groupby('CID').agg({
    'TID': 'count',                      # Number of transactions (frequency)
    'Net Amount': 'sum',                 # Total purchase value
    'Location': 'first',                 # City of the customer
    'Gender': 'first'                    # Gender of the customer
}).reset_index()

# Rename columns for clarity
customer_loyalty.columns = ['Customer ID', 'Purchase Frequency', 'Total Spend', 'Location', 'Gender']

# Define thresholds or scores for loyalty and filter the most loyal customers
def loyalty_score(row):
    if row['Purchase Frequency'] > 3 and row['Total Spend'] > 5000:
        return 'High'
    elif row['Purchase Frequency'] > 1 and row['Total Spend'] > 2000:
        return 'Medium'
    else:
        return 'Low'

# Apply the loyalty score function
customer_loyalty['Loyalty Score'] = customer_loyalty.apply(loyalty_score, axis=1)

# Filter for high loyalty customers and those with gender as 'Others'
loyal_customers_others = customer_loyalty[
    (customer_loyalty['Loyalty Score'] == 'High') &
    (customer_loyalty['Gender'] == 'Others')
]

# Count the number of loyal customers by city
loyalty_by_city = loyal_customers_others['Location'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(10, 6))
plt.pie(loyalty_by_city, labels=loyalty_by_city.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("YlGnBu", len(loyalty_by_city)))
plt.title('Customer Loyalty Distribution by City for Gender "Others"')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()

# Optional: Display the loyalty count by city DataFrame
print(loyalty_by_city)

# Optional: Export the results to a CSV file
loyalty_by_city.to_csv('loyal_customers_others_by_city.csv', index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (replace with your actual dataset path)
data = pd.read_csv("/content/project datapool_df.csv")

# Print column names to verify the correct names for filtering
print(data.columns)

# Check for the correct column names for gender and discount amount
if 'Gender' not in data.columns or 'Discount Amount (INR)' not in data.columns or 'Location' not in data.columns:
    raise ValueError("The dataset must contain 'Gender', 'Discount Amount (INR)', and 'Location' columns.")

# Group by gender and city, summing the discount amounts
gender_city_discount = data.groupby(['Location', 'Gender'])['Discount Amount (INR)'].sum().reset_index()

# Debugging output to check the sum of discounts
print(gender_city_discount)

# Create pie charts for each gender
genders = ['Female', 'Male', 'Other']
plt.figure(figsize=(18, 6))

for i, gender in enumerate(genders):
    plt.subplot(1, 3, i + 1)

    # Filter data for the current gender
    gender_data = gender_city_discount[gender_city_discount['Gender'] == gender]

    # Plot pie chart
    plt.pie(gender_data['Discount Amount (INR)'],
            labels=gender_data['Location'],
            autopct='%1.1f%%',
            startangle=140)
    plt.title(f'Total Discount Amount for {gender} Customers by City')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
